```python
#K-means
#Question 1

import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse.linalg import svds
array=np.genfromtxt('MovieRankings36.csv',delimiter=',')
row=[0,1,3,7,15,24,28]
col=[1,4,3,5,10,6,11,13,17,0]
arr=array[row,:][:,col]
k=3
rep_vec=np.random.rand(1,len(col),k)*5
prev_rep=np.zeros((1,len(col),k))
onesPos=np.array(np.random.choice(k, len(row)))
L=np.eye(len(row),k)[onesPos]
arrtemp=np.dstack((arr,arr))
arr=np.dstack((arrtemp,arr))
print("rep_vec R:")
print(rep_vec)
RNorm_list=[np.linalg.norm(arr[:,:,0]-np.matmul(L,np.transpose(rep_vec.reshape(len(col),k))))]

while(np.linalg.norm(rep_vec-prev_rep)>1e-12):

    b=np.argmin(np.linalg.norm((arr-rep_vec),axis=1),axis=1)
    b=b.reshape(len(row),1)
    L=np.zeros((len(row),k))
    np.put_along_axis(L,b,1,axis=1)
    prev_rep=rep_vec.copy()
    #print("here")
    for x in range(k):
        j=np.argwhere(b==x)
        j=j[:,0]
        if(j.size!=0):
            rep_vec[0,:,x]=np.mean(arr[j,:,0],axis=0)
        else:
            rep_vec[0,:,x]=np.random.rand(1,len(col))*5
    RNorm_list.append(np.linalg.norm(arr[:,:,0]-np.matmul(L,np.transpose(rep_vec.reshape(len(col),k)))))
```

```python

#PGD
A=arr[:,:,0]
W=A[:,:k]
H=np.hstack((np.eye(k),np.zeros((k,len(col)-k))))
alpha=0.01

R=A-np.matmul(W,H)
R_list=[np.linalg.norm(R, 'fro')]
iter_max=1
while(iter_max<1000):
    
    W=W+alpha*np.matmul(R,np.transpose(H))
    #print(W)
    W[W<0]=0
    H=H+alpha*np.matmul(np.transpose(W),R)
    H[H<0]=0
    R=A-np.matmul(W,H)
    R_list.append(np.linalg.norm(R, 'fro'))
    iter_max+=1
    #print(W)
    #print(H)
```

```python
plt.plot(R_list)
plt.ylabel('frobenius norm')
plt.xlabel("iterations")
plt.title("PGD")
plt.show()
```

```python
# Lee Sung
A=arr[:,:,0]
W=A[:,:k]
H=np.hstack((np.eye(k),np.zeros((k,len(col)-k))))
stepW=np.ones((len(row),k))*0.01
stepH=np.ones((k,len(col)))*0.001

R=A-np.matmul(W,H)
R_listLee=[np.linalg.norm(R, 'fro')]
iter_max=1
while(iter_max<2000):
    
    W=W+np.multiply(stepW,np.matmul(R,np.transpose(H)))
    #print(W)
    W[W<0]=0
    H=H+np.multiply(stepH,np.matmul(np.transpose(W),R))
    H[H<0]=0
    R=A-np.matmul(W,H)
    R_listLee.append(np.linalg.norm(R, 'fro'))
    iter_max+=1
    #print(W)
    #print(H)
```

```python
plt.plot(R_listLee)
plt.ylabel('frobenius norm')
plt.xlabel("iterations")
plt.title("Lee and Sung")
plt.show()
```

```python
#Question 2
#Low rank Approximation (Alternating Iteration)\
lambdaa=[0.1,0.01,0.05]
for l in lambdaa:
    A=array
    A=np.nan_to_num(A,nan=0)
    #u, s, v = np.linalg.svd(A, full_matrices=False)
    X=np.random.rand(A.shape[0],k)
    Y=np.random.rand(A.shape[1],k)
    
    k=3
    n=A.shape[0]
    d=A.shape[1]
    R=A-np.matmul(X,Y.T)
    R_AlternativeIteration=[np.linalg.norm(R, 'fro')]


    for iter_max in range(10):
        for i in range(n):
            X[i,:]=np.linalg.solve(Y.T @ Y + l*np.eye(k),Y.T @ A[i,:].T).T
        for i in range(d):
            Y[i,:]=np.linalg.solve(X.T @ X + l*np.eye(k),X.T @ A[:,i].T).T
        R=A-np.matmul(X,Y.T)
        R_AlternativeIteration.append(np.linalg.norm(R, 'fro'))

    plt.plot(R_AlternativeIteration)
    plt.ylabel('frobenius norm')
    plt.xlabel("iterations")
    plt.legend(["lam1","lam2","lam3"])
    plt.title("Alternating iterations")
    print(min(R_AlternativeIteration))
plt.show()
```

```python
#Question 2(b)
#Nuclear Norm trick


lambdaa=[1.01,5,10]
k=3
A=array
i=0
for l in lambdaa:
    A=np.nan_to_num(A,nan=0)
    #u, s, v = np.linalg.svd(A, full_matrices=False)
    M=np.random.rand(A.shape[0],A.shape[1])

    n=A.shape[0]
    d=A.shape[1]
    R=A-M
    R_AlternativeIteration=[np.linalg.norm(R, 'fro')]
    colours=['r','g','b']
    

    for iter_max in range(10):
        u,s,v=np.linalg.svd(np.nan_to_num(A-M,nan=0)+M,full_matrices=False)
        s=np.diag(s)
        s[s<l]=0
        M=u @ s @ v
        R=A-M
        R_AlternativeIteration.append(np.linalg.norm(R, 'fro'))

    plt.plot(R_AlternativeIteration,colours[i])
    i+=1
    print(min(R_AlternativeIteration))
plt.ylabel('frobenius norm')
plt.xlabel("iterations")
plt.legend(["lam1","lam2","lam3"])
plt.title("Nuclear norm Trick")
plt.show()


```

```python
#Question 3
#CUR decomposition

import time
from statistics import mean 
A_mat=np.genfromtxt('M.csv',delimiter=',')

def Col_Select(M,c,k):
    u,s,v=svds(M,k)
    #v=v[:,:k]
    p=np.square(np.linalg.norm(v,axis=1))*(1/k)
    ptemp=p*c
    eta=np.random.rand(ptemp.shape[0],1)
    C=M[:,list(np.argwhere(eta<ptemp).reshape(-1))]
    return C


#M.shape
MeanNorm=[]
MeanRatioNorm=np.zeros((8,9))
for a in range (1,9):
    for k in range (2,11):
        c=a*k
        u, s, vt = svds(A_mat, k)
        A_matK=(u*s) @ vt
        if a==1:
            MeanNorm.append(np.linalg.norm((A_mat-A_matK),'fro'))
        t0=time.clock()
        MeanRatio=np.zeros((100))
        for iters in range(100):
            C=Col_Select(A_mat,c,k)
            R=Col_Select(A_mat.T,c,k)
            U=np.linalg.pinv(C) @ A_mat @ np.linalg.pinv(R.T)
            MeanRatio[iters]=np.linalg.norm((A_mat-(C @ U @ R.T)),'fro')/np.linalg.norm((A_mat-A_matK),'fro')
        #print(time.time()-t0)    
        MeanRatioNorm[a-1][k-2]=(np.mean(MeanRatio))
        
    #plt.plot(MeanRatioNorm)
print(MeanRatioNorm)  
print(MeanRatioNorm.shape)

nx, ny = (8, 9)
x = np.linspace(0, 8, ny)

y = np.linspace(0, 7, nx)

X,Y=np.meshgrid(x, y)

plt.contourf(X, Y, MeanRatioNorm)
plt.title("MeanRatioNorm")
plt.colorbar()
plt.show()  

plt.plot(MeanNorm)
plt.title("Mean SVDk difference Norm")
plt.show() 
```
